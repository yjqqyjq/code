```python
def circle(a,b,x0,N):
  theta = np.random.uniform(0,2*np.pi,size=N)
  r = 1 * np.sqrt(np.random.uniform(0, 1,size=N))
  x = x0 + a*r * np.cos(theta)
  y =  b*r * np.sin(theta)
  return np.array([x,y])

points1=circle(1,1,-1,5000)
points2=circle(1,1,1,5000)
points=np.concatenate((points1,points2),axis=1)
points2=circle(2,1,0,10000)
q1=fn.quadrupole(points[0],points[1],np.zeros(10000),10000)
q2=fn.quadrupole(points2[0],points2[1],np.zeros(10000),10000)
print(q1,q2)
```
