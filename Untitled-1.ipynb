```python
import h5py
import numpy as np
import matplotlib.pyplot as plt
f=h5py.File('/Users/24756376/data/Flamingo/L1000N0900/halo_xray.hdf5','r')
Coord=np.array(f["PartType0"]['Coordinates'])
Xray=np.array(f["PartType0"]['xray_lum_erosita_high'])
f.close()
com=Coord.mean(axis=0)
Coord-=com
r=np.sqrt((Coord[:,0])**2+(Coord[:,1])**2+(Coord[:,2])**2)
bins=np.linspace(0,2,15)
bin=0.5*(bins[1:]+bins[:-1])
rho=np.histogram(r,bins=bins)[0]/bin**2
rho_xray=np.histogram(r,weights=Xray,bins=bins)[0]/bin**2
fig=plt.figure()
ax=fig.add_subplot(1,1,1)
ax.plot(bin,rho/rho[0],label='gas density profile')
ax.plot(bin,rho_xray/rho_xray[0],label='X-ray luminosity profile')


ax.set_yscale('log')
ax.legend()
ax.set_xlabel(r"$r[Mpc]$")

ax.set_ylabel(r"$\rho/\rho_{c}$")
```
