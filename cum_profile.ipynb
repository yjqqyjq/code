```python
import matplotlib.pyplot as plt
import numpy as np
import h5py
from scipy.interpolate import interp1d
f=h5py.File("/Users/24756376/data/Flamingo/L1000N1800/cum_profiles",'r')
bin=np.array(f["r"])
hb=np.array(f["baryons"])
hdm=np.array(f["dm"])
f.close()
f=h5py.File("/Users/24756376/data/Flamingo/L1000N1800_NoCool/cum_profiles",'r')
hb_ad=np.array(f["baryons"])
hdm_ad=np.array(f["dm"])
f.close()
r75_dm=np.zeros(len(hb))
r75_b=np.zeros(len(hb))
for i in range(len(hb)):
    
    hdm[i]=hdm[i]/hdm[:,8][i]
    hb[i]=hb[i]/hb[:,8][i]
    mass_profile=interp1d(hdm[i],bin)
    r75_dm[i]=mass_profile(0.25)
    mass_profile=interp1d(hb[i],bin)
    r75_b[i]=mass_profile(0.25)
   
    hdm_ad[i]=hdm_ad[i]/hdm_ad[:,8][i]
    hb_ad[i]=hb_ad[i]/hb_ad[:,8][i]
f=h5py.File("/Users/24756376/data/Flamingo/L1000N1800/c_z_half.hdf5",'a')
f.create_dataset("cdm_50_25",data=r75_dm)
f.create_dataset("cb_50_25",data=r75_b)
f.close()    
'''
fig=plt.figure()
ax=fig.add_subplot(1,1,1)
hg=hb*8.56/(hdm*45.2+hb*8.56)/(0.0494/0.316)
hg_ad=hb_ad*8.56/(hdm_ad*45.2+hb_ad*8.56)/(0.0494/0.316)
ax.plot(bin,np.percentile(hg_ad,50,axis=0),label='Ad')
ax.plot(bin,np.percentile(hg,50,axis=0),label='Hydro')
ax.set_xscale('log')

#ax.fill_between(bin,np.percentile(hg,25,axis=0),np.percentile(hg,75,axis=0),alpha=0.3)
'''
```
