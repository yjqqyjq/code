```python
import matplotlib.pyplot as plt
import numpy as np
import h5py
f=h5py.File("/Users/24756376/data/Flamingo/L1000N1800/halos_ranked.hdf5",'r')

id=np.array(f["id"])
#input_id=np.array(f["input_ids"])[id<=0]
centers=np.array([f["centers_x"],f["centers_y"],f["centers_z"]]).T[id<=0]
id=-id[id<=0]
f.close()
f=h5py.File("/Users/24756376/data/Flamingo/L1000N1800_Nocool/halos_ranked.hdf5",'r')
id_ad=np.array(f["id"])

#input_id_ad=np.array(f["input_ids"])[id_ad<=0]
centers_ad=np.array([f["centers_x"],f["centers_y"],f["centers_z"]]).T[id_ad<=0]
id_ad=-id_ad[id_ad<=0]
f.close()
f=h5py.File("/Users/24756376/data/Flamingo/L1000N1800/same_halos.hdf5",'r')
same_id=np.array(f["ids"])
f.close()
f=h5py.File("/Users/24756376/data/Flamingo/L1000N1800_NoCool/same_halos.hdf5",'r')
same_id_ad=np.array(f["ids"])
f.close()
centers_final=centers[same_id]
centers_final_ad=centers_ad[same_id_ad]
fig = plt.figure()
ax=plt.subplot(1,1,1)
ax.scatter(centers_final[:,0][0:100],centers_final[:,2][0:100],s=0.5,color="r",label="Hydro",alpha=0.5)
ax.scatter(centers_final_ad[:,0][0:100],centers_final_ad[:,2][0:100],s=0.5,color="b",label="Ad",alpha=0.5)
ax.legend()
```
