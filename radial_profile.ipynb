```python
import matplotlib.pyplot as plt
import numpy as np
import h5py
f=h5py.File("/Users/24756376/data/Flamingo/L1000N1800_NoCool/radial_profile",'r')
hdm=np.array(f['Ndm'])
hg=np.array(f['Ng'])

f.close()
bins=10**np.linspace(-1.5,np.log10(2.5),25)
bin=10**((np.linspace(-1.5,np.log10(2.5),25)[1:]+np.linspace(-1.5,np.log10(2.5),25)[:-1])/2)
per=np.percentile((hg*8.56)/(hdm*45.2+hg*8.56)/((0.0494/0.316)),[25,50,75],axis=0)
fig=plt.figure()
ax=fig.add_subplot(111)
print(np.log10(bins))
ax.plot(bin,per[1],label='Adiabatic')
ax.fill_between(bin,per[0],per[2],alpha=0.3)
ax.set_xscale('log')


f=h5py.File("/Users/24756376/data/Flamingo/L1000N1800/radial_profile.hdf5",'r')
hdm=np.array(f['Ndm'])
hg=np.array(f['Ng'])
hs=np.array(f['Ns'])
f.close()
bins=10**np.linspace(-1.5,np.log10(2.5),25)
bin=10**((np.linspace(-1.5,np.log10(2.5),25)[1:]+np.linspace(-1.5,np.log10(2.5),25)[:-1])/2)
per=np.percentile(hg/hdm,[25,50,75],axis=0)
#fig=plt.figure()
#ax=fig.add_subplot(111)
#print(np.log10(bins))
ax.plot(bin,per[1],label='Hydro')
ax.fill_between(bin,per[0],per[2],alpha=0.3)
ax.set_xscale('log')
ax.legend()
ax.set_xlabel('r/r50_c')
ax.set_ylabel('f_b/fb_ave ')
```
